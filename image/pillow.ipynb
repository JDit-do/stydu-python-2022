## 2.1 파일 열기 및 옵션 사용
사용 내용 정리
 - glob: 패턴 및 경로에 해당하는 파일 찾기.
 - Image.open: 이미지 파일 열어서, Image 객체 로드해서 다양한 처리나 변환 작업 등 수행 가능
    - e.g.) resize, rotate, filter, convert 등
 - ImageFilter: 이미지에 여러 효과 적용하는데 사용.

````python
from glob import glob
from PIL import Image, ImageFilter

fileList = glob('./lib/img/*.jpg') # 패턴에 해당하는 파일 가져옴.
file = fileList[0]

# 파일 확인
imgExample1 = Image.open(file)
# 관련 옵션
imgExample1.resize((400, 300))
imgExample1.rotate(45)
imgExample1.filter(ImageFilter.CONTOUR) # 윤곽선
imgExample1.convert('L') # 이미지 흑백으로 변환
````

## 2.2 파일이 없을 때

````python
from glob import glob
from PIL import Image
import csv

def imgOpen(file):
    try:
        baseFolder = './lib/img/'
        file = baseFolder + file
        img = Image.open(file)
        print(img.format, img.size, img.mode)
    except IOError as err:  
        print("%s unable to load image"%file)

with open('./lib/csv/imgList.csv', newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter='\t')
    for row in csvreader:
            print(row)
            imgOpen(row[0])
    
````
