```python
# 데이터

speechList = ['''Ladies and Gentlemen,
As we gather here today, we are faced with an undeniable truth that our planet is undergoing profound changes. The climate crisis is not a distant threat; it is here, it is now, and it demands our immediate attention.
For decades, human activities have led to an unprecedented rise in greenhouse gas emissions, resulting in rising global temperatures, melting ice caps, and extreme weather events. The consequences are clear: communities displaced, ecosystems destroyed, and vulnerable populations left even more exposed.
But let us not be paralyzed by despair. Instead, let us harness the power of innovation, collaboration, and determination to forge a path forward. It is our moral obligation, not just to ourselves but to future generations, to act swiftly and decisively.
We must transition to renewable energy sources, promote sustainable practices, and invest in resilient infrastructure. Governments, businesses, and individuals alike must come together, setting aside differences and prioritizing the health and well-being of our planet.
Inaction is no longer an option. The time for excuses has passed. We must seize this moment, united in purpose and resolve, to safeguard our planet for ourselves and for all who will come after us.
Together, we can make a difference. Together, we can create a sustainable future for all.
Thank you.''', '''Esteemed Colleagues and Friends,
As we gather today, let us recognize the urgency of the climate crisis facing our world. The consequences of inaction are dire, threatening communities, economies, and ecosystems alike.
We must transition to sustainable practices, embrace innovation, and prioritize environmental conservation. Our planet's future depends on the choices we make today.
Let us commit to a greener, more sustainable future, working hand in hand to address the challenges of climate change. Thank you.
''', '''Dear Citizens and Leaders,
The alarm bells of climate change ring louder with each passing day. From the scorched lands to the rising sea levels, our Earth is sending distress signals that demand our immediate attention.
We cannot afford to be bystanders. It is time to prioritize sustainability over convenience, long-term well-being over short-term gains. Let us embrace renewable energies, adopt eco-friendly practices, and preserve our natural habitats.
By taking bold actions today, we pave the way for a brighter, healthier tomorrow for all living beings on this planet. Together, let's be the change-makers our world desperately needs. Thank you.''']
```

```python
import matplotlib.pyplot as plt

from collections import Counter
from wordcloud import WordCloud

for speechText in speechList:
    lowerText = speechText.lower()
    words = lowerText.split()

    # Counter: 주어진 데이터에서 각 항목의 빈도를 계산
    wordFreq = Counter(words)
    wordCloud = WordCloud(width=800, height=400, background_color='white').generate(speechText)

    # 생성되 그림의 크기 지정.
    plt.figure(figsize=(10, 5))
    plt.imshow(wordCloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```
